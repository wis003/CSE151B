## Install the package dependencies before running this notebook

```python
import torch
from torch.utils.data import Dataset, DataLoader
import os, os.path 
import numpy 
import pickle
from glob import glob

"""
    number of trajectories in each city
    # austin --  train: 43041 test: 6325 
    # miami -- train: 55029 test:7971
    # pittsburgh -- train: 43544 test: 6361
    # dearborn -- train: 24465 test: 3671
    # washington-dc -- train: 25744 test: 3829
    # palo-alto -- train:  11993 test:1686

    trajectories sampled at 10HZ rate, input 5 seconds, output 6 seconds
    
"""
```

## Create a Torch.Dataset class for the training dataset

```python
from glob import glob
import pickle
import numpy as np

ROOT_PATH = "./"

cities = ["austin", "miami", "pittsburgh", "dearborn", "washington-dc", "palo-alto"]
splits = ["train", "test"]

def get_city_trajectories(city="palo-alto", split="train", normalized=False):
    f_in = ROOT_PATH + split + "/" + city + "_inputs"
    inputs = pickle.load(open(f_in, "rb"))
    inputs = np.asarray(inputs)
    
    outputs = None
    
    if split=="train":
        f_out = ROOT_PATH + split + "/" + city + "_outputs"
        outputs = pickle.load(open(f_out, "rb"))
        outputs = np.asarray(outputs)

    return inputs, outputs

class ArgoverseDataset(Dataset):
    """Dataset class for Argoverse"""
    def __init__(self, city: str, split:str, transform=None):
        super(ArgoverseDataset, self).__init__()
        self.transform = transform

        self.inputs, self.outputs = get_city_trajectories(city=city, split=split, normalized=False)

    def __len__(self):
        return len(self.inputs)

    def __getitem__(self, idx):

        data = (self.inputs[idx], self.outputs[idx])
            
        if self.transform:
            data = self.transform(data)

        return data
```

## Sample a batch of data and visualize 

```python
import matplotlib.pyplot as plt
import random

def show_sample_batch(sample_batch, version):
    """visualize the trajectory for a batch of samples"""
    inp, out = sample_batch
    batch_sz = inp.size(0)
    agent_sz = inp.size(1)
    
    fig, axs = plt.subplots(1,batch_sz, figsize=(15, 3), facecolor='w', edgecolor='k')
    fig.subplots_adjust(hspace = .5, wspace=.001)
#     axs.set_title(version)
    axs = axs.ravel()
    
    for i in range(batch_sz):
        axs[i].xaxis.set_ticks([])
        axs[i].yaxis.set_ticks([])
        
        # first two feature dimensions are (x,y) positions
        axs[i].scatter(inp[i,:,0], inp[i,:,1])
        axs[i].scatter(out[i,:,0], out[i,:,1])

city = 'austin'
train_dataset = ArgoverseDataset(city = city, split = 'train')
train_loader = DataLoader(train_dataset, batch_size=4)      
for i_batch, sample_batch in enumerate(train_loader):
    if i_batch % 3000 == 0:
        inp, out = sample_batch
        print(inp.shape)
        print(out.shape)
        show_sample_batch(sample_batch, str(i_batch) + '_' + city)
```

## Heatmaps

```python
def heat_map(x, y, title):
    print(np.amin(x))
    print(np.amin(y))
    x += -1 * np.amin(x)
    y += -1 * np.amin(y)
    hm = np.zeros((int(np.amax(x) / 10) + 1, int(np.amax(y) / 10) + 1))
    print(hm.shape)
    for i in range(len(x)):
        hm[int(x[i] / 10), int(y[i] / 10)] += 1 # occurrences
    
    plt.figure(figsize=(8, 8))
    plt.imshow(hm, cmap='nipy_spectral_r', vmin=0, vmax=1)
    plt.xlabel('X coordinates')
    plt.ylabel('Y coordinates')
    plt.title(title, fontweight ="bold")
    plt.show()   
    

test_dataset = ArgoverseDataset(city = 'austin', split = 'train')
total_input = test_dataset.inputs
total_output = test_dataset.outputs
for city in cities:
    print(city)
    test_dataset = ArgoverseDataset(city = city, split = 'train')
    
    x = np.concatenate((np.ravel(test_dataset.inputs[:, :, 0]), np.ravel(test_dataset.outputs[:, :, 0])))
    y = np.concatenate((np.ravel(test_dataset.inputs[:, :, 1]), np.ravel(test_dataset.outputs[:, :, 1])))
    heat_map(x, y, city)
    
    if city != 'austin':
        total_input = np.concatenate((total_input, test_dataset.inputs))
        total_output = np.concatenate((total_output, test_dataset.outputs))

# expect (~203816, 50, 2), (~203816, 60, 2)
print("Total training data:")
print(total_input.shape)
print(total_output.shape)
```
